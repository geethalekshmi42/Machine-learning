5.HANDLE OUTLIERS 

```python
#IQR method for outlier removal
def remove_outliers(df,columns):
  data_filtered=df.copy()

  for column in columns:
     Q1 = df[column].quantile(0.25)
     Q3 = df[column].quantile(0.75)
     IQR = Q3 - Q1

     lower_bound = Q1 - 1.5 * IQR
     upper_bound = Q3 + 1.5* IQR

     data_filtered = data_filtered[(data_filtered[column] >= lower_bound) & (data_filtered[column] <= upper_bound)]

  return data_filtered
```

```python
new_data=remove_outliers(df,['Votes','Hondt','FinalMandates','totalMandates','availableMandates','numParishes'])
```

```python
new_data=remove_outliers(df,['numParishes','numParishesApproved','blankVotes','nullVotes','subscribedVoters','totalVoters'])
```
